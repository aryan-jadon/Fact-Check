```python
def calculate_automated_readability_index(article_text):
    try:
        doc = make_spacy_doc(article_text, lang="en_core_web_sm")
        article_information = text_stats.api.TextStats(doc)
        return article_information.automated_readability_index
    except Exception as e:
        return 0.0

def calculate_coleman_liau_index(article_text):
    try:
        doc = make_spacy_doc(article_text, lang="en_core_web_sm")
        article_information = text_stats.api.TextStats(doc)
        return article_information.coleman_liau_index
    except Exception as e:
        return 0.0

def calculate_flesch_kincaid_grade_level(article_text):
    try:
        doc = make_spacy_doc(article_text, lang="en_core_web_sm")
        article_information = text_stats.api.TextStats(doc)
        return article_information.flesch_kincaid_grade_level
    except Exception as e:
        return 0.0

def calculate_flesch_reading_ease(article_text):
    try:
        doc = make_spacy_doc(article_text, lang="en_core_web_sm")
        article_information = text_stats.api.TextStats(doc)
        return article_information.flesch_reading_ease
    except Exception as e:
        return 0.0

def calculate_gunning_fog_index(article_text):
    try:
        doc = make_spacy_doc(article_text, lang="en_core_web_sm")
        article_information = text_stats.api.TextStats(doc)
        return article_information.gunning_fog_index
    except Exception as e:
        return 0.0

def calculate_smog_index(article_text):
    try:
        doc = make_spacy_doc(article_text, lang="en_core_web_sm")
        article_information = text_stats.api.TextStats(doc)
        return article_information.smog_index
    except Exception as e:
        return 0.0
```

```python
cnn_articles = pd.read_excel("Files/CNN_Articles.xlsx")

# calculating micro factor 1
cnn_articles['readability']=cnn_articles['ArticleText'].progress_apply(lambda x: calculate_automated_readability_index(x))

# calculating micro factor 2
cnn_articles['understanding']=cnn_articles['ArticleText'].progress_apply(lambda x: calculate_coleman_liau_index(x))

# calculating micro factor 3
cnn_articles['grade']=cnn_articles['ArticleText'].progress_apply(lambda x: calculate_flesch_kincaid_grade_level(x))

# calculating micro factor 4
cnn_articles['reading_ease']=cnn_articles['ArticleText'].progress_apply(lambda x: calculate_flesch_reading_ease(x))

# calculating micro factor 5
cnn_articles['education_factor_1']=cnn_articles['ArticleText'].progress_apply(lambda x: calculate_gunning_fog_index(x))

# calculating micro factor 6
cnn_articles['education_factor_2']=cnn_articles['ArticleText'].progress_apply(lambda x: calculate_smog_index(x))

cnn_articles.to_excel(file_path,index=False)
processed_cnn_articles = cnn_articles
```
